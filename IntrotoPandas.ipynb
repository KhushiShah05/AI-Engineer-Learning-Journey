A Series is single column of data,


```python
import pandas as pd
s = pd.Series([10,20,30,40])
print(s)
```

Custom Index

```python
marks = pd.Series([86,76,89],
                  index = ["Maths","English","Science"])
print(marks)
```

Accessing Values

```python
s = pd.Series([100,200,300])
print(s[2])
print(s.dtype)
```
