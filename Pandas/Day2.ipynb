Dataframe


```python
import pandas as pd
data = {
    "Name": ["Khushi","Rahul","Riya"],
    "Age" : [20,21,19]
}
df = pd.DataFrame(data)
print(df)

```

Custom Index


```python
import pandas as pd

data = {
    "Name": ["Khushi","Rahul","Riya"]
}
df = pd.DataFrame(
    data,
    index=["A","B","C"]
)
print(df)
print(df.shape)
print(df.columns)
print(df.index)
print(df.size)
print(df["Name"])
```

Add new column


```python
import pandas as pd
data = {
    "Name":["Khushi","Rahul"]
}
df = pd.DataFrame(data)
df["Age"] = [20,21]
print(df)
```
